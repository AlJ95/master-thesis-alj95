```python
import json
data = json.load(open("../../../data/PyCommitsWithDiffs.json"))
print(data.keys())
```

```python
entry = data['https://github.com/mTresk/android_kernel_oneplus_sdm845']
entry
```

```python
import pandas as pd
repo, values = next(iter(data.items()))
commit_id, commit_values = next(iter(values.items()))
commit_values.update({"repo": repo, "commit_id": commit_id})
pd.DataFrame([commit_values])
```

```python
import pandas as pd
dfs = []
for repo, values in data.items():
    for commit_id, commit_values in values.items():
        commit_values.update({"repo": repo, "commit_id": commit_id})
        dfs.append(
            pd.DataFrame([commit_values])
        )
data_df = pd.concat(dfs)
data_df
```
